```python
# Common parames
p = 0.5
h = 0.5
pM = p*h
c = 4.0
# BTC params
uB = 12.5
# FTC params
c1 = 0.1
u1 = uB*c1
u2 = (1-c1)*uB*h
pO = (1-h)*pM
pN = 1-pM
```

```python
import math
def nCr(n,r):
    f = math.factorial
    return f(n) / f(r) / f(n-r)
```

```python
def getGainBTC(r):
    return ((1-pM)**(r-1))*pM*uB
def getCostBTC(r):
    return ((1-pM)**(r-1))*pM*(c**r)
```

```python
def getGainFTC(r):
    gain = 0
    for k in range(1,r):
        gain += nCr(r-1, k) * (pO**k) * (pN**(r-1-k)) * (u1 + (k+1)*u2)
    gain *= pM
    return gain
def getCostFTC(r):
    return ((1-pM)**(r-1))*pM*(c**r)
```

```python
gainLstBTC, costLstBTC, profitLstBTC = [], [], []
gainLstFTC, costLstFTC, profitLstFTC = [], [], []
```

```python
for r in range(1,145):
    gainBTC, costBTC = 0, 0
    gainFTC, costFTC = 0, 0
    
    for i in range(1, r):
        gainBTC += getGainBTC(i)
        costBTC += getCostBTC(i)
        gainFTC += getGainFTC(i)
        costFTC += getCostFTC(i)
        
    gainLstBTC.append(gainBTC)
    costLstBTC.append(costBTC)
    profitLstBTC.append(gainBTC-costBTC)
    
    gainLstFTC.append(gainFTC)
    costLstFTC.append(costFTC)
    profitLstFTC.append(gainFTC-costFTC)
```
