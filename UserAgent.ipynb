```python
I = 100
f = 20
pF = 0.5
cF = 2.0
```

```python
def gain(r):
    return ((1-pF)**(r-1))*pF*(I-f)
```

```python
def cost(r):
    return ((1-pF)**(r-1))*cF**(r-1)
```

```python
gainLog = []
costLog = []
profitLog = []
for i in range(1, 145):
    sumGain = 0
    sumCost = 0
    for r in range(1, i):
        sumGain += gain(r)
        sumCost += cost(r)
    gainLog.append(sumGain)
    costLog.append(sumCost)
    profitLog.append(sumGain-sumCost)
```
